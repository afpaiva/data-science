Importações das libs:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from scipy.stats import pearsonr
```

- 1) Ler o dataset "*Advertising/Advertising.csv*" como dataframe utilizando a biblioteca Pandas e mostrar as 5 primeiras e 5 últimas linhas.

```python
df = pd.read_csv('/content/DSBD/Datasets/Advertising/Advertising.csv')
df = df.dropna()
df
```

- 2) Mostre num mesmo quadro (usando subplots) os três diagramas de dispersão para os pares de variáveis: TV x sales, radio x sales e newspaper x sales.

```python
plt.subplot(1,3,1) 
sbn.scatterplot(x=df['TV'],y=df['sales'])
plt.subplot(1,3,2) 
sbn.scatterplot(x=df['radio'],y=df['sales'])
plt.subplot(1,3,3) 
sbn.scatterplot(x=df['newspaper'],y=df['sales'])

plt.subplots_adjust(left=0, right=2.0, wspace=.3, hspace=.5)
plt.show()
```

- 3) Calcule o coeficiente de correlação de Pearson para os três pares de variáveis: TV x sales, radio x sales e newspaper x sales.

```python
print('Coeficiente de correlação de Pearson:\n')
r, _ = pearsonr(df['TV'], df['sales'])
print('TV x Sales:\t\t', round(r, 2))
r, _ = pearsonr(df['radio'], df['sales'])
print('Radio x Sales:\t\t', round(r, 2))
r, _ = pearsonr(df['newspaper'], df['sales'])
print('Newspaper x Sales:\t', round(r, 2))
```

- 4) Analisando os diagramas de dispersão acima e os valores dos coeficientes de correlação responda qual dos tipos de anúncios influencia mais e qual deles influencia menos nos resultados das vendas?

O gráfico de dispersão TV/vendas é o que melhor representa uma linearidade e menor dispersão entre os pontos. Significa que há uma forte correlação entre TV e vendas, por isso o valor do Coef. de correlação de Pearson foi o mais próximo de 1.
Já o de dispersão Newspaper/vendas apresentou uma dispersão maior e impossibilidade de identificação de uma linearidade entre os pontos. O Coef. de correlação de Pearson foi baixo, 0.23, logo há uma relação muito fraca entre Newspaper e vendas.
